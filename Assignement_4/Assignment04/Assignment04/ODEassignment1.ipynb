# Solving ODEs and plotting
$$\dot{y} = - \lambda y$$


```python
lam = 2           # lambda
dt  = 0.05        # time step
T   = 0.2         # total integration time 
nt  = round(T/dt) # total number of time steps
y   = 1           # initial conditions 
y_values = [y]    # list to store y values for plotting

# (1a) Write a function here
def ordinary_ODE(lam, dt, y):
    return y - dt * lam * y

for it in range(0, nt):
    y = ordinary_ODE(lam, dt, y)
    y_values.append(y)
    print("t_n = %-.2f, y_n = %-.3f" % ((it + 1) * dt, y))  

# Plotting command (1b)
import matplotlib.pyplot as plt

time_points = [i * dt for i in range(nt + 1)]
plt.plot(time_points, y_values, 'gs') 
plt.xlabel('Time')
plt.ylabel('y')
plt.title('Solution of Ordinary Differential Equation')
plt.grid(True)
plt.show()
```

# 2 Backward Euler method
The following code uses the backward Euler method to solve the ODE$$\dot{y} = - \lambda y.$$ In the following questions, you will modify this code to solve the ODE $$\dot{y} = (y-3)(y+1).$$

```python
def backward_euler(y, dt):
    return (y - 3) * (y + 1) * dt

dt = 0.05        # time step
T = 2            # total integration time
nt = round(T / dt) # total number of time steps
y = 2            # initial condition
y_values = [y]   # list to store y values for plotting

for it in range(nt):
    y = (y + backward_euler(y, dt)) / (1 + backward_euler(y, dt))
    y_values.append(y)
    print("t_n = %-.2f, y_n = %-.3f" % ((it + 1) * dt, y))  # Print values for demonstration

import matplotlib.pyplot as plt

time_points = [i * dt for i in range(nt + 1)]
plt.plot(time_points, y_values, 'gs')  # Plotting points as green squares
plt.xlabel('Time')
plt.ylabel('y')
plt.title('Solution of Ordinary Differential Equation using Backward Euler')
plt.grid(True)
plt.show()
```

## b)
Using the backward Euler method with Newton iterations, plot the solution for the initial conditions $y_0 = 2$ and with a time step of $\Delta t = 0.05$.  

```python
def f(y):
    return (y - 3) * (y + 1)

def f_prime(y):
    return 2*y - 2

def g(y_next):
    return y_next - 3

def g_prime(y_next):
    return 1 + f_prime(y_next)

def backward_euler_newton(y, dt, max_iter=100, tol=1e-6):
    y_next = y
    for _ in range(max_iter):
        y_next_new = y + f(y_next) * dt + g(y_next) * dt
        if abs(y_next_new - y_next) < tol:
            return y_next_new
        y_next = y_next_new
    return None

dt = 0.05        # time step
T = 2            # total integration time
nt = round(T / dt) # total number of time steps
y = 2            # initial condition
y_values = [y]   # list to store y values for plotting

for it in range(nt):
    y = backward_euler_newton(y, dt)
    y_values.append(y)
    print("t_n = %-.2f, y_n = %-.3f" % ((it + 1) * dt, y))  # Print values for demonstration

import matplotlib.pyplot as plt

time_points = [i * dt for i in range(nt + 1)]
plt.plot(time_points, y_values, 'gs')  # Plotting points as green squares
plt.xlabel('Time')
plt.ylabel('y')
plt.title('Solution of Ordinary Differential Equation using Backward Euler with Newton Iterations')
plt.grid(True)
plt.show()
```
